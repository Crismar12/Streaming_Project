### Importaciones de las librerías

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pyarrow
import os
```

### Top 5 géneros más populares por plataforma

```python
top5_por_plataforma = pd.read_parquet(r"C:\Users\FRANCIS\Documents\Proyecto_Streaming\data\top5_por_plataforma.parquet", engine="pyarrow")
plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=top5_por_plataforma,
    x="listed_in",
    y="conteo",
    hue="plataforma",
    dodge=True,
    palette="Set2"
)
plt.xticks(rotation=30, ha="right")
plt.title("Top 5 géneros más populares por plataforma")
plt.xlabel("Género")
plt.ylabel("Cantidad de títulos")
plt.tight_layout()

# Agregar etiquetas de valor encima de cada barra
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2, height), 
                    ha='center', va='bottom', fontsize=9, color='black', rotation=0)

plt.show()
```

### ¿Cuántas películas y series se añaden a la plataforma cada año?

```python
df_anual = pd.read_parquet(r"C:\Users\FRANCIS\Documents\Proyecto_Streaming\data\df_anual.parquet", engine="pyarrow")

g = sns.FacetGrid(
    df_anual,
    col="type",
    height=5,
    aspect=1.5,
    sharey=True
)

g.map_dataframe(
    sns.barplot,
    x="year_added",
    y="conteo",
    hue="plataforma",
    palette="Set2"
)

for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', label_type='edge', padding=2, fontsize=7)

g.add_legend(title="Plataforma")
g.set_axis_labels("Año", "Cantidad de títulos añadidos")
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.85)
g.fig.suptitle("Películas y series añadidas por año y plataforma")
plt.show()
```
